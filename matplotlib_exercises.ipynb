```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
x = np.linspace(-2, 5, 100)
y = x ** 2 - x + 2
 
fig = plt.figure(figsize = (10, 5))

plt.plot(x, y)
plt.title('Problem #1')
plt.show()
```

Create and label 4 separate charts with the following equations (choose a range for x that makes sense):

y=√x

y=x3

y=2x

y=1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
xv = np.array([0, 1, 2, 3, 4])
yv = np.zeros(len(xv))


print('--------------')
print('x      |    y  ')
print('--------------')
for i in range(len(xv)):
    yv[i] = np.sqrt(xv[i])
    print('{:5.2f}  | {:5.2f}'.format(xv[i],yv[i]))

print('--------------')
```

```python
plt.plot(xv, yv)
plt.title('y=√x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y = x ** 3
 
fig = plt.figure(figsize = (10, 5))
plt.plot(x, y)
plt.title('y=x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.array(range(-10,10))#(-2, 2, 100)
y = 2 * x
 
fig = plt.figure(figsize = (10, 5))
plt.plot(x, y)
plt.title('y=2x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
